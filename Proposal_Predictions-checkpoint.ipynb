```python
## Choosing rows
data_to_use = data[data.Stage.isin(['Closed Won', 'Closed Lost'])]
data_to_use = data_to_use[-data_to_use['Actual B&P Amt'].isnull()]
data_to_use = data_to_use[-data_to_use['Difficulty Level'].isnull()]
data_to_use['Actual B&P Amt'] = data_to_use['Actual B&P Amt'].apply(lambda x: float(str(x).replace('(', '').replace(')', '').replace(',', '').split('.')[0]))
#data_to_use = data_to_use.dropna()

## Creating the variables
columns_to_drop = ['Opportunity ID', 'Proposal #', 'Opportunity Name']

boolean_cols = ['Competitive (Y/N)', 'Veej is Incumbent (Y/N)', 'Incumbents (Y/N)',
                'Add-On Project (Y/N)', 'Task Order under Existing IDIQ (Y/N)', 'Veej Prime (Y/N)',
                'Subcontractors (Y/N)']

for col in boolean_cols:
    data_to_use = data_to_use[data_to_use[col].isin(['Y', 'N'])]
    data_to_use[col] = data_to_use[col] == 'Y'

    
columns_to_use  = ['Difficulty Level', 'Amount', 'Actual B&P Amt'] + boolean_cols


columns_to_get_dummy = ['Contract Type', 'Contract Type Category', ]

# Scaling the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data_to_use[columns_to_use])

X = pd.DataFrame(scaler.transform(data_to_use[columns_to_use]), columns=columns_to_use) #data_to_use[columns_to_use]
y = data_to_use['Stage']
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import roc_curve, auc

import pydotplus
from itertools import cycle
```

```python
# Models
rf_model = RandomForestClassifier()
dt_model = DecisionTreeClassifier(max_depth=3)
lg_model = LogisticRegression(solver = 'lbfgs')


# Compute ROC curve and ROC area
fpr = dict()
tpr = dict()
roc_auc = dict()


classifiers = {'Decision Tree'       : dt_model,
               'Logistic Regression' : lg_model,
               'Random Forest'       : rf_model
              }

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

for name, model in classifiers.items():
    print(name)
    model.fit(X_train, y_train)
    #y_preds = model.predict_proba(X_test)[:,1]
    y_preds = model.predict(X_test)
    fpr[name], tpr[name], _ = roc_curve(y_test.values.ravel() == 'Closed Won', y_preds == 'Closed Won')
    #fpr[name], tpr[name], _ = roc_curve(y_test.values.ravel() == 'Closed Won', y_preds == 'Closed Won')
    roc_auc[name] = auc(fpr[name], tpr[name])

plt.figure()
lw = 2

colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
for model_name, color in zip(classifiers.keys(), colors):
    plt.plot(fpr[model_name], tpr[model_name], color=color, lw=lw,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(model_name, roc_auc[model_name]))

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('1 - False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve - Proposal predictions')
plt.legend(loc="lower right")
plt.show()
```

```python
# Predictions
predictions_df = pd.DataFrame(list(zip(y_test,y_preds)), columns = ['Real value','Predicted value'])
print(sum(predictions_df['Real value'] == predictions_df['Predicted value']) / predictions_df.shape[0])
predictions_df.head()
```
